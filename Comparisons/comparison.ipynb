<h3> Loading all saved values for BTSD Dataset on Alexnet <h3>

```python
PATH = '/Users/vishalkarmakar/Downloads/Alex_BTSD_Full.pt'
model = torch.hub.load('pytorch/vision:v0.10.0', 'alexnet', pretrained=False)
model.eval()
checkpoint = torch.load(PATH,map_location='cpu')
model = model.load_state_dict(checkpoint['model_state_dict'])
accuracy_list = checkpoint['accuracy']
loss_list = checkpoint['loss']
```

<h3> CNN Model Comparison of Alexnet vs Googlenet vs VGG16 vs Resnet on BTSD Datset <h3>

```python
pyplot.figure(figsize=(8, 8))
pyplot.plot(np.array(accuracy_list)) #Add list of Alexnet here
#pyplot.plot(np.array(accuracy_list)) #Add list of vgg here
#pyplot.plot(np.array(accuracy_list2)) #Add list of googlenet here
#pyplot.plot(np.array(accuracy_list2)) #Add list of resnet here

# Using Plt.scatter, visualize the transformed data marked with their corresponding label in 2D embedding.
font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'black','size':15}

pyplot.title("Alexnet vs Googlenet vs VGG16 vs Resnet", fontdict=font1)
pyplot.xlabel("Steps(*100)", fontdict=font2)
pyplot.ylabel("Accuracies (%)", fontdict=font2)

pyplot.legend(["Accuracy of Alexnet(BTSD)", "Accuracy of Model 2"], loc ="upper right")
pyplot.show()
```
